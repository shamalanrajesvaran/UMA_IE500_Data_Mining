```python
import pandas as pd
import numpy as np

# Load the data
data = pd.read_csv(r'C:\Users\cheng\Documents\GitHub\Data_Mining_Project\UMA_IE500_Data_Mining\hotel_booking_data\hotel_bookings.csv')

# Replace 'NULL' string values with real NaN
data.replace('NULL', np.nan, inplace=True)

# Convert date columns to datetime format
data['arrival_date'] = pd.to_datetime(
    data['arrival_date_year'].astype(str) + '-' +
    data['arrival_date_month'].astype(str) + '-' +
    data['arrival_date_day_of_month'].astype(str),
    errors='coerce'
)

# Drop redundant date parts (optional)
# data.drop(['arrival_date_year', 'arrival_date_month', 'arrival_date_day_of_month'], axis=1, inplace=True)

# Convert categorical and numeric columns
categorical_cols = data.select_dtypes(include='object').columns
numeric_cols = data.select_dtypes(include='number').columns

# Optional: fix boolean style numeric values
data['is_canceled'] = data['is_canceled'].astype(int)

# Overview of cleaned data
print(data.info())
print("\nMissing values per column:\n", data.isnull().sum())
print("\nSample rows:\n", data.head())
```
