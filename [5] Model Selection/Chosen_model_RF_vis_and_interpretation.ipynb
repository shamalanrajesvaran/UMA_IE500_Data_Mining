# This notebook is specifically design for visualizing and analysing the selected best suitable model - Random Forest

```python
import joblib
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
import pandas as pd

# Load the saved model
rf_model = joblib.load(r"C:\Users\cheng\Documents\GitHub\UMA_IE500_Data_Mining\Model_training\random_forest_best_model.pkl")

# Extract the feature names from the training data
train_data = pd.read_csv(r"C:\Users\cheng\Documents\GitHub\UMA_IE500_Data_Mining\[3] full dataset preparation and baseline model evaluation\train_data.csv")  # or however you originally loaded it
X_train = train_data.drop('is_canceled', axis=1)
feature_names = X_train.columns.tolist()

# Extract one decision tree
tree = rf_model.estimators_[0]

# Plot the tree
plt.figure(figsize=(20, 10))
plot_tree(tree, 
          filled=True, 
          feature_names=feature_names, 
          max_depth=2, 
          fontsize=13)  # or 16

plt.title("One Decision Tree from Random Forest")
plt.show()
```
